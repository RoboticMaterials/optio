# Use Mongo Data 

```python
client = MongoClient('localhost:27017')
db = client.ContactDB
skills_collection = db.skills

skills = skills_collection.find()
skill_list = list(skills_collection.find())
condition_list = []
new_condition_list = []

for skill in skills:
    condition_list.extend(skill['postconditions'])

for condition in condition_list:
    if condition not in new_condition_list and condition['type'] != 'stack':
        new_condition_list.append(condition)

for condition in new_condition_list:
    print(json.dumps(condition))
```

## Get task list

```python
client = MongoClient('localhost:27017')
db = client.ContactDB
skills_collection = db.skills
tasks_collection = db.tasks
skill_list = list(skills_collection.find())
# skill_list = skills_collection.find()
task_list = list(tasks_collection.find())
for task in task_list:
    print(task['name'], task['_id'])
```

## Find task dict and convert to task

```python
task_id = '5ec8076f38dcd9bc3b460d07'
task_found = False
for task in task_list:
    if str(task['_id']) == task_id:
        task_dict = task
        task_found = True
if not task_found: print('Task Not Found')

pprint(task_dict)
    
task = rmengine.task.Task()
task.convert_dict_to_task(task_dict)
task.print_task()

pprint(task.convert_task_to_dict())
```
